First, we want to mention that the matrix indices are course indices (such as `0, 1, 2, 3`). Very often we need to query the actual course ids (such as `ML0151EN` and `ML0101ENv3`) based on course indices and vice versa. We can save the course id's and indices into two dictionaries for late queries:


Then, based on the `doc_index` and `doc_id` columns, we create an index to id mapping and another id to index mapping in two Python dictionaries:


````python
# Create course id to index and index to id mappings
def get_doc_dicts(bow_df):
    # Group the DataFrame by course index and ID, and get the maximum value for each group
    grouped_df = bow_df.groupby(['doc_index', 'doc_id']).max().reset_index(drop=False)
    # Create a dictionary mapping indices to course IDs
    idx_id_dict = grouped_df[['doc_id']].to_dict()['doc_id']
    # Create a dictionary mapping course IDs to indices
    id_idx_dict = {v: k for k, v in idx_id_dict.items()}
    # Clean up temporary DataFrame
    del grouped_df
    return idx_id_dict, id_idx_dict
````

Now, you can iterate each unselect course and check if it is similar enough to any of your selected courses. If the similarity is larger than a threshold such as 0.5 or 0.6, then add it to your course recommendation list:


_TODO: Complete the following method to recommend courses which are similar to your enrolled courses_


````python
def generate_recommendations_for_one_user(enrolled_course_ids, unselected_course_ids, id_idx_dict, sim_matrix):
    """
    Generates a list of recommended courses for a single user based on item-item similarity.
    """
    # Create a dictionary to store potential recommendations.
    # Key: unselected course ID, Value: highest similarity score found so far.
    res = {}
    threshold = 0.5

    # Iterate over each course the user is already enrolled in.
    for enrolled_course in enrolled_course_ids:
        # Check if the enrolled course is in our dictionary to avoid errors.
        if enrolled_course not in id_idx_dict:
            continue

        # Get the index for the enrolled course.
        idx_1 = id_idx_dict[enrolled_course]

        # Iterate over courses the user has NOT taken.
        for unselect_course in unselected_course_ids:
            # Check if the unselected course is in our dictionary.
            if unselect_course not in id_idx_dict:
                continue

            # Get the index for the unselected course.
            idx_2 = id_idx_dict[unselect_course]
            
            # Get the similarity score from the matrix.
            sim = sim_matrix[idx_1][idx_2]

            # If the similarity is above the threshold, consider it for recommendation.
            if sim > threshold:
                # If the course is not yet in our results, add it.
                # If it is already there, update its score only if the new score is higher.
                current_score = res.get(unselect_course, 0)
                if sim > current_score:
                    res[unselect_course] = sim

    # Sort the final recommendations by score in descending order.
    res = {k: v for k, v in sorted(res.items(), key=lambda item: item[1], reverse=True)}
    
    # Limit the final recommendations to the top 20.
    top_20_res = dict(list(res.items())[:20])
    
    return top_20_res
````

````python
# Assume these variables are defined and loaded from your project environment
# sim_url, course_url, bow_url, test_users_df, id_idx_dict, sim_matrix
# and the function generate_recommendations_for_one_user()

def generate_recommendations_for_all():
    """
    Generates recommendations for all test users, skipping any user
    who has no valid recommendations.
    """
    users = []
    courses = []
    sim_scores = []
    
    # Load necessary data (placeholders for your actual loading code)
    # sim_df = pd.read_csv(sim_url)
    # course_df = pd.read_csv(course_url)
    # all_courses = set(course_df['COURSE_ID'])
    # bow_df = pd.read_csv(bow_url)
    
    # Prepare the list of test users
    test_users = test_users_df.groupby(['user']).max().reset_index(drop=False)
    test_user_ids = test_users['user'].to_list()
    
    # Loop through each user to generate recommendations
    for user_id in test_user_ids:
        # Get courses the user has already enrolled in
        enrolled_course_ids = test_users[test_users['user'] == user_id]['item']
        unselected_course_ids = all_courses.difference(enrolled_course_ids)
        
        # Generate recommendations for this single user
        res = generate_recommendations_for_one_user(
            enrolled_course_ids, 
            unselected_course_ids, 
            id_idx_dict, 
            sim_matrix
        )
        
        # Process the results into temporary lists
        cv = []
        sc = []
        # Check if 'res' is a dictionary before trying to call .items()
        if isinstance(res, dict):
            for k, v in res.items():
                if k is not None:
                    cv.append(k)
                    sc.append(v)
        
        # --- CORRECTED LOGIC ---
        # Only if the processed course list (cv) is NOT empty,
        # do we add the user and their recommendations to the final lists.
        if cv:  # This is a concise way to check if a list is not empty
            users.append(user_id)
            courses.append(cv)
            sim_scores.append(sc)
            
    return users, courses, sim_scores
````
